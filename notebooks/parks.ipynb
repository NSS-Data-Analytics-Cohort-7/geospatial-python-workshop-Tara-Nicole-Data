```python
center = zipcodes.geometry.centroid[12]
print(center)
```

```python
county_center = [center.y, center.x]
```

```python
print(county_center)
```

```python
map_county = folium.Map(location = county_center, zoom_start = 12)
map_county
```

```python
i = 0
for row_index, row_values in parks_by_zip.iterrows():
    if i <3:
        print('index is', row_index)
        print('values are:')
        print(' ')
        print(row_values)
        print('------------------------- ')
        i+=1
```

```python
map_county = folium.Map(location = county_center, zoom_start = 12)

folium.GeoJson(parks_by_zip).add_to(map_county)

for row_index, row_values in parks_by_zip.iterrows():
    loc = [row_values['zip'], [row_values['geometry']]
    pop = str(row_values['common_nam'])
    icon=folium.Icon(color="darkgreen",icon="tree", prefix='fa')
    
    marker = folium.Marker(
        location = loc, 
        popup = pop, icon = icon) 

    marker.add_to(map_county)
map_county.save('../maps/mapcounty.html')

map_county
```
